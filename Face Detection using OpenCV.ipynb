```python
import numpy as np
import pandas as pd
import cv2
from matplotlib import pyplot as plt
import glob
```

```python
txtfiles = [] 
for file in glob.glob("*.jpg",recursive=True):
    txtfiles.append(file)
```

```python

for ix in txtfiles:
    img = cv2.imread(ix,cv2.IMREAD_COLOR)
    imgtest1 = img.copy()
    imgtest = cv2.cvtColor(imgtest1, cv2.COLOR_BGR2GRAY)
    facecascade = cv2.CascadeClassifier('C:/Users/manju/Documents/Term II/Machine Learning/haarcascade_frontalface_default.xml')
    eye_cascade = cv2.CascadeClassifier('C:/Users/manju/Documents/Term II/Machine Learning/haarcascade_eye.xml')
    faces = facecascade.detectMultiScale(imgtest, scaleFactor=1.2, minNeighbors=5)
    print('Total number of Faces found',len(faces))
    for item,(x, y,w,h) in enumerate(faces):
        face_detect = cv2.rectangle(imgtest, (x, y), (x+w, y+h), (200, 0, 200), 2)
        roi_gray = imgtest[y:y+h, x:x+w]
        roi_color = imgtest[y:y+h, x:x+w]        
        plt.imshow(face_detect)
    plt.show()
        
```
